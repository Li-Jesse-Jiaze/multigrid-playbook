```python
import numpy as np
from matplotlib import pyplot as plt

n = 129
omega = 2/3
diag = 2 * np.ones(n)
off = -1 * np.ones(n - 1)
A = np.diag(diag) + np.diag(off, 1) + np.diag(off, -1)
S = np.eye(n) - (omega / 2.0) * A
m = (n + 1) // 2
P = np.zeros((n, m))
for j in range(m):
    i = 2 * j
    P[i, j] = 1.0
    if i - 1 >= 0:
        P[i - 1, j] = 0.5
    if i + 1 < n:
        P[i + 1, j] = 0.5
A_c = P.T @ A @ P
T = np.eye(n) - P @ np.linalg.inv(A_c) @ P.T @ A
ST = S @ T
plt.plot(np.sort(np.abs(np.real_if_close(np.linalg.eigvals(ST)))), label=r"$|\lambda(ST)|$")
plt.plot([0, n], [1/3, 1/3], label=r"$1/3$")
plt.legend()
plt.ylim(0, 1)
plt.show()
```
